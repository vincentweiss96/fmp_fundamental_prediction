```python
import pandas as pd
```

```python
import FundamentalAnalysis as fa
api_key = "key"
```

```python
#comparison values
ticker="AAPL"

balance_sheet = fa.balance_sheet_statement("AAPL", api_key, period="annual")
income_statement = fa.income_statement("AAPL", api_key, period="annual")
cash_flow_statement = fa.cash_flow_statement("AAPL", api_key, period="annual")

IS = income_statement.transpose()
CF = cash_flow_statement.transpose()
BS = balance_sheet.transpose()
ratios = fa.financial_ratios("AAPL", api_key, period="annual").transpose()

IS_compare=IS.columns.values
CF_compare=CF.columns.values
BS_compare=BS.columns.values
ratios_compare=ratios.columns.values
```

```python
companies=pd.read_csv("Data\\FMP\\FMP_tickers.csv")
Concatenate=pd.read_csv("FMP_Concatenate.csv")
```

```python
Concatenate.dropna(subset=["fillingDate"], inplace=True)
```

```python
companies.set_index("Unnamed: 0", inplace=True)
companies.index.name = None
companies.reset_index(drop=True, inplace = True)
Concatenate.set_index("Unnamed: 0", drop = True, inplace = True)
Concatenate.index.name = None
```

```python
Concatenate.rename(columns={"fillingDate.1": "fillingDate", \
                            "acceptedDate.1": "acceptedDate",\
                           "period.1":"period",\
                            "inventory.1": "inventory",\
                            'link.1':'link',\
                            'finalLink.1':'finalLink',\
                            'fillingDate.2':'fillingDate',\
                            'acceptedDate.2':'acceptedDate',\
                            'period.2':'period',\
                            'depreciationAndAmortization.1':'depreciationAndAmortization',\
                            'netIncome.1':'netIncome',\
                            'link.2':'link',\
                            'finalLink.2':'finalLink'}, inplace = True)
```

```python
skipped_list=[]
IS_compare_l=31
CF_compare_l=35
BS_compare_l=44
ratios_compare_l=54
```

```python
i=22000
while i <22500:
    
    ticker=companies["symbol"][i]
    print(i)
    i+=1
    
    if not ticker in Concatenate['Ticker'].unique():
                
        try:
            
            balance_sheet = fa.balance_sheet_statement(ticker, api_key, period="annual")
            income_statement = fa.income_statement(ticker, api_key, period="annual")
            cash_flow_statement = fa.cash_flow_statement(ticker, api_key, period="annual")
            IS = income_statement.transpose()
            CF = cash_flow_statement.transpose()
            BS = balance_sheet.transpose()
            ratios = fa.financial_ratios(ticker, api_key, period="annual").transpose()
            


            if IS_compare_l!=len(IS.columns.values):
                for col in IS_compare:
                    try:
                        IS.insert(0,col,"")
                    except Exception as e:
                        print(e)
                        continue
            if CF_compare_l!=len(CF.columns.values):
                for col in CF_compare:
                    try:
                        CF.insert(0,col,"")
                    except Exception as e:
                        print(e)
                        continue
            if BS_compare_l!=len(BS.columns.values):
                for col in BS_compare:
                    try:
                        BS.insert(0,col,"")
                    except Exception as e:
                        print(e)
                        continue
            if ratios_compare_l!=len(ratios.columns.values):
                for col in ratios_compare:
                    try:
                        ratios.insert(0,col,"")
                    except Exception as e:
                        print(e)
                        continue

            
            Concatenate_this = pd.concat([BS, CF, IS, ratios], axis=1, sort=False)
            Concatenate_this.insert(0,"Ticker",ticker)
            Concatenate_this.reset_index(inplace=True)

            Concatenate = Concatenate.append(Concatenate_this, ignore_index=True, sort=False)
            
        except Exception as e:
            
            skipped_list.append(ticker)
            
            
    else:
        print (ticker + " already in list")
```

```python
Concatenate.to_csv("FMP_Concatenate.csv")
#companies.to_csv("Data\\FMP\\FMP_tickers.csv")
```
